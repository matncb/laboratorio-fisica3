```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots

plt.style.use(['science', 'notebook', 'grid'])
```

```python
r_anel = np.array([2,1,1,1,1,1,1,2])*1e-2
V_anel = np.array([0.26, 0.10, 0.01, 0.00, 0.01, -0.01, 0.22, 0.29])
x_anel = np.array([3.0, 4.5, 5.5, 6.5, 7.5, 8.5, 9.5, 11.0])

r_placa = np.array([2,2,2,2]) *1e-2
V_placa = np.array([0.21,0.20,0.21,0.23])
x_placa = np.array([4,6,8,10])
```

```python
E_anel = V_anel/r_anel
E_anel
```

```python
E_anel_err = 0.04/r_anel + np.abs(V_anel)/(r_anel**2) *0.002
E_anel_err
```

```python
E_placa = V_placa/r_placa
E_placa
```

```python
E_placa_err = 0.04/r_placa + np.abs(V_placa)/(r_placa**2) *0.002
E_placa_err
```

```python
plt.errorbar(x_placa, E_placa, xerr=0.2, yerr=E_placa_err, marker = 'o', ls = '', label = 'Placa')
plt.errorbar(x_anel, E_anel, xerr=0.2, yerr=E_anel_err, marker = 'o', ls = '', label = "Anel")

plt.xlabel("x (cm)")
plt.ylabel("E (V/m)")

plt.legend()
plt.title("Campo elétrico em função da distância")
```
